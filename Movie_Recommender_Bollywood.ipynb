```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',token_pattern=r'\b\w+\b')
```

```python
vector=cv.fit_transform(df['tags'])
```

```python
vector.toarray()
```

```python
len(vector.toarray()[0])
```

```python
newdf=df[['title','tags','poster_path']]
```

```python
newdf.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
print(similarity)
```

```python
def recommend(movie):
    movie_index=newdf[newdf['title']==movie].index[0]
#     print(movie_index)
    distances=similarity[movie_index]
#     print(distances)
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
#     print(movies_list)
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
```

```python
d=input("Enter movie to recommend: ")
print("Recommendations based on your choices are: ")
recommend(d)
```
